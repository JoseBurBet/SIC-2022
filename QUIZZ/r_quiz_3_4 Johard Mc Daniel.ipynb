Defina una función llamada is_palindrome y escriba un programa que reciba una cadena del usuario e imprima si el palíndromo es correcto o no. Pauta de código: Llame a la función is_palindrome dentro de la función is_palindrome(función recursiva).

```python

def es_palindrome(palabra):
        if len(palabra) < 1:
            return True
        else:
            if palabra[0] == palabra[-1]:
                return es_palindrome(palabra[1:len(palabra)-1])
            else:
                return False
        
chain = input("Introduzca alguna palabra: ")

if es_palindrome(chain) is True:
    print(f'La palabra "{chain}" es un palíndromo')
else:
    print(f'La palabra "{chain}" no es un palíndromo')

    
```
