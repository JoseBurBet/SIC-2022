```python
from time import gmtime, localtime,  strftime

print(localtime()) #gmt/utc -5
fecha_hora=strftime("%d/%m/%Y %H:%M:%S",localtime())
print()
print(fecha_hora)
print(type(fecha_hora))

fecha=strftime("%d-%m-%Y",localtime())
print()
print(fecha)

hora=strftime("%H:%M:%S",localtime()) #por defecto viene en formato 24h
print()
print(hora)

hora=strftime("%I:%M:%S %p",localtime())# formato 12h
print()
print(hora)


gm=strftime("%d/%m/%Y %H:%M:%S",gmtime()) #gmt/utc 00
print(gm)

```

# PANDAS
### pandas.series
### pandas.dataframe

```python
# NO cuento el index
#series ->  solo tengo una columna
#dataframe -> puedo tener1 o mas columnas

import pandas as pd

# tengo una tienda de ropa

#si tengo una columna uso series

productos=["camisetas","jeans","sacos","zapatos"]
inventario=[7,5,2,9]

serie1=pd.Series(inventario) #(datos,index)
print(serie1)
print("________________________")
serie1=pd.Series(inventario,productos) #(datos,index)
print(serie1)

dic_tienda=dict(zip(productos,inventario)) # convierto 2 listas en un diccionario

print(dic_tienda)
print("\n diccionario \n")
serie2=pd.Series(dic_tienda)
print(serie2)

#lelgo mi jefe y me puso precios
precios=[20,30,50,40]

# cuanto tengo mas columnas paso a dataframes

#creo un diccionario de series

dic_1={"productos":productos,"inventario":inventario,"precios":precios}
print(dic_1)

df=pd.DataFrame(dic_1)
df
```

```python
# recopilando solo dataframe

productos=["camisetas","jeans","sacos","zapatos"]
inventario=[7,5,2,9]
precios=[20,30,50,40]

dic_1={"productos":productos,"inventario":inventario,"precios":precios}
print(dic_1)

df=pd.DataFrame(dic_1)
print(df)

from IPython.core.display import HTML
display(HTML(df.to_html())) # esto reemplaza nuestro print para ver los df bonitos

df.set_index("productos",inplace=True)
display(HTML(df.to_html()))
```
