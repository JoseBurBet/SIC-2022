```python
import time
print(time.ctime())
print(type(time.ctime()))# me da el tiempo en string

print(time.localtime()) # la hora local es la hora de donde se encuentre la maquina
print(type(time.localtime()))
```

```python
#import time
#time.localtime()

#time la usaremos cuando necesitamos solo mostrar tiempos

from time import gmtime,localtime,strftime

print(localtime())# la hora del computador
fecha_hora=strftime("%d-%m-%Y %H:%M:%S",localtime()) #(formato,tiempo)
print()
print(fecha_hora)

fecha=strftime("%d-%m-%Y",localtime()) #(formato,tiempo)
print(fecha)

hora=strftime("%H:%M:%S",localtime()) #formato 24h
print(hora)
print(type(hora))
print()
hora=strftime("%I:%M:%S %p",localtime()) #formato 12h
print(hora)

fecha=strftime("%d/%m/%Y",localtime()) #(formato,tiempo) #GMT/UTC -5
print(fecha)

gmhora=strftime("%I:%M:%S %p",gmtime()) #formato 12h #GMT/UTC 00
print(gmhora)

print("__________________")
horamilitar=strftime("%H%M%S",localtime()) #formato 24h
print(horamilitar)
militar=int(horamilitar)
print(type(militar))
```

```python
# cronometro
import time

for i in range(10):
    time.sleep(1) #delay->espera un tiempo en segundos
    print(strftime("%I:%M:%S %p",localtime())) #formato 12h
 
```

```python
#reloj
import time

while True:
    time.sleep(1) #delay->espera un tiempo en segundos
    print(strftime("%I:%M:%S %p",localtime())) #formato 12h
```

## Datetime


```python
# datetime la usaremos cuando necesitemos OPERAR tiempos

from datetime import datetime

actual=datetime.now()
print(actual)
print(type(actual))

print(actual.hour)
print(actual.month)
```

```python
#ejemplo

#necesito reportar a china a que hora salio el barco del puerto en al hora de ellos

from datetime import datetime, timedelta
#timedelta permite operar entre tiempos

panama=datetime.now()
print("la hora en panama es: ",panama) #GMT/UTC -5

##GMT/UTC en china? = #GMT/UTC +8

#diferencia son 13h

diferencia=timedelta(hours=13) # convertir el 13 a formato de tiempo
#print(type(diferencia))
hora_china=panama+diferencia
print("la hora en china  es: ",hora_china) #GMT/UTC +8
```

```python
# NO cuento el index, index no es una columna

#series-> 1 columna
#dataframe-> 1  o mas columnas

import pandas as pd

#trabajo para una tienda de ropa

#si tengo 1 columna entocnes series

productos=["camisetas","faldas","jeans","sombreros","medias"]
inventario=[6,7,3,9,2]

serie1=pd.Series(inventario)
print(serie1)

serie2=pd.Series(inventario,productos) #(datos,index)
print(serie2)
```

```python
productos=["camisetas","faldas","jeans","sombreros","medias"]
inventario=[6,7,3,9,2]

#dic={"key":value,"key2":value2}
#dic={"camisetas":6,"faldas":7......}

dic=dict(zip(productos,inventario))
print(dic)

serie3=pd.Series(dic)
print(serie3)

#llega el jefe y me pide tener una columna de precios

precios=[20,15,30,20,5]

# como tengo ahora mas de 1 columna paso a dataframes

#construyo un diccionario DE SERIES!!!!!!

dic_tienda={"productos":productos,"inventario":inventario,"precios":precios}
df=pd.DataFrame(dic_tienda)
#print(df)

from IPython.core.display import HTML
display(HTML(df.to_html())) # esto reemplaza nuestro print para ver los df bonitos


df.set_index("productos",inplace=True)
display(HTML(df.to_html()))
```
