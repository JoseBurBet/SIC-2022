```python
from time import gmtime, strftime ,localtime

gmtime()
print(localtime())
print(type(localtime()))
print()
fecha_hora=strftime("%d-%m-%Y %H:%M:%S",localtime())
print(fecha_hora)
print(type(fecha_hora))

fecha=strftime("%d-%m-%Y",localtime())
print(fecha)

Hora=strftime("%H:%M:%S",localtime()) #GMT/UTC -5 panama
print(Hora)

HoraGM=strftime("%H:%M:%S",gmtime()) #formato 24h
print(HoraGM)

HoraGM=strftime("%I:%M:%S %p",gmtime()) #formato 12h
print(HoraGM) #hora del meridiano greenwich GMT/utc +00 
```

# Datetime 

```python
from datetime import datetime

ahora=datetime.now()

print(ahora)
print(type(ahora))

print()

hora_nueva=ahora.replace(hour=7)
print(hora_nueva)
# saca como datetime y en orden

print("___________________")

print(ahora)
print(ahora.day)
print(ahora.hour)

```

# Libreria Pandas

```python
#series -> es un dataframe de 1 sola columna
#dataframes -> un excel

import pandas as pd

#tengo una tienda de ropa


# si tengo una sola columna uso series

productos=["camisas","pantalones","sombreros","zapatos"]
ventas=[4,8,7,4]


s1=pd.Series(ventas,productos) #datos,indice
s1

# si tengo mas de 1 columna uso dataframe

precios=[20,25,10,30]

#creo un diccionario de series
dic_datos={"productos":productos,"ventas":ventas,"precios":precios}

#convierto el diccionario en dataframe
dt=pd.DataFrame(dic_datos)


from IPython.core.display import HTML
display(HTML(dt.to_html()))

dt.set_index("productos",inplace=True) # movimos la columna date para que ahora sea el index
display(HTML(dt.to_html()))
```
