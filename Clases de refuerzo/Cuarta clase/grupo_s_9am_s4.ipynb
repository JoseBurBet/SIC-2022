```python
#import time
from time import gmtime,localtime,strftime

print(localtime()) #la hora local del computador
fecha_hora=strftime("%d-%m-%Y %H:%M:%S",localtime()) #(formato,hora)
print("\nla hora local es: ",fecha_hora)

fecha=strftime("%d-%m-%Y",localtime())
print(fecha)

hora=strftime("%H:%M:%S",localtime()) # aqui tenemos formato 24h
print(hora)

hora=strftime("%I:%M:%S %p",localtime()) # aqui tenemos formato 24h
print(hora)
print(type(hora))

print("__________________________")
print(strftime("%I:%M:%S %p",localtime()))# GMT/UTC -5
print(strftime("%I:%M:%S %p",gmtime()))   # GMT/UTC 00
```

```python
#cronometro
import time
for i in range(10):
    time.sleep(1) #sleep espera una cantidad en segundos
    print(strftime("%I:%M:%S %p",localtime()))
```

### Datetime

```python
#datetime la usaremos cuando necesito operar horas

from datetime import datetime

actual=datetime.now() #imprime bonito,tipo datetime
print("la hora en panama es: ",actual)
print(type(actual))
```

```python
#ejemplo 

#neccesito reportar a la empresa en china a que hora 
#salio mi barco desde panama, en la hora de ellos

from datetime import datetime, timedelta
#timedelta me permite operar entre tiempos

actual=datetime.now() 
print("la hora en panama es: ",actual) # GMT/UTC -5

#gmt de china? =   GMT/UTC +8

#diferencia horaria son 13 h

diferencia=timedelta(hours=13) #convertimos el numero 13 en formato de tiempo
hora_china=actual+diferencia #sumamos
print("la hora en china es : ",hora_china) # GMT/UTC +8
```

# Pandas
### pandas.Series
### pandas.DataFrame

```python
#NO cuento el index, index no es columna

#series -> solo tengo una columna
#dataframe -> una o mass columnas

import pandas as pd

# tengo una tienda de ropa

#si tengo 1 columna uso series

productos=["camisetas","zapatos","abrigos","faldas"]
inventario=[8,10,12,6]

serie1=pd.Series(inventario)
print(serie1)
print()
serie2=pd.Series(inventario,productos)#(values,index)
print(serie2)
```

```python
productos=["camisetas","zapatos","abrigos","faldas"]
inventario=[8,10,12,6]

#dic_tienda={"camisetas":8,"zapatos":10} # NO a mano
dic_tienda=dict(zip(productos,inventario))
print(dic_tienda)

serie3=pd.Series(dic_tienda)
print(serie3)


#llega mi jefe y me dice que necesita poner otra columna

precios=[20,30,35,20]


#uso dataframes

#construyo un diccionario de series

dic_completo={"productos":productos,"inventario":inventario,"precios":precios}
df=pd.DataFrame(dic_completo)
print(df)
```

```python
#ejem completo dataframe

productos=["camisetas","zapatos","abrigos","faldas"]
inventario=[8,10,12,6]
precios=[20,30,35,20]

dic_completo={"productos":productos,"inventario":inventario,"precios":precios}
df=pd.DataFrame(dic_completo)

from IPython.core.display import HTML
display(HTML(df.to_html())) # esto reemplaza nuestro print para ver los df bonitos

df.set_index("productos",inplace=True)
display(HTML(df.to_html()))
```
