## 1) importar los datos a python

```python
# normalmente los archivos csv vienen separados por , este es especial y esta separado por ;

import pandas as pd

data_mat=pd.read_csv("student-mat.csv",sep=";") #data original
data_por=pd.read_csv("student-por.csv",sep=";") #data original

data_mat
```
