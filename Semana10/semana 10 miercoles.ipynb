# Multi-indexing pandas dataframe

```python
#EJEMPLO 1 ->MANUAL

import pandas as pd
from IPython.core.display import HTML

dic1={
    "year":[2018,2018,2018,2019,2019,2019],
    "country":["spain","portugal","france","spain","portugal","france"]
     }
df=pd.DataFrame(dic1)
display(HTML(df.to_html()))

index=pd.MultiIndex.from_frame(df)

data=pd.DataFrame(data=[1234,5678,9012,3456,7890,1237],index=index,columns=["vaccinations"])

display(HTML(data.to_html()))


```

```python
#EJEMPLO 1 ->GROUPBY

import pandas as pd
from IPython.core.display import HTML
dic2={
    "year":[2018,2018,2018,2019,2019,2019],
    "country":["spain","portugal","france","spain","portugal","france"],
    "vaccinations":[1234,5678,9012,3456,7890,1237]
     }
df2=pd.DataFrame(dic2)
display(HTML(df2.to_html()))

df_g_y=df2.groupby(["year"])

for key,group in df_g_y:
    
    print("+key ", key)
    print("+number ", len(group))
    print(group.head())
    print("\n")
    
    
print("____________________________________________________________________________")    

df_g_c=df2.groupby(["country"])

for key,group in df_g_c:
    
    print("+key ", key)
    print("+number ", len(group))
    print(group.head())
    print("\n")
```

```python
#https://seaborn.pydata.org
#https://github.com/mwaskom/seaborn-data    

import seaborn as sns
from IPython.core.display import HTML

#print(sns.get_dataset_names()) # para saber que bases de datos hay disponibles en el git de seaborn


df=sns.load_dataset("titanic") # esta e sla manera para traer un archivo desde el repositorio de Github
#display(HTML(df.to_html()))
print("el total de filas es :", len(df))
print()
print(df.isnull().sum(axis=0))

```

# ejemplo1 ELIMINAR los valores NaN

```python
print("\n datos originales \n ")   
print(df.isnull().sum(axis=0))

nan_deck=df["deck"].value_counts(dropna=False) #check si hay nan en la columna deck
#print(nan_deck)

#________________________ELIMINAR POR COLUMNAS_______________________________

#la columna no es representativa , la elimino
df_new=df.dropna(axis=1,thresh=500)#dropna elimina valores nulos, axis 1 indica columnas
                            # thresh 500 indica que borre las columnas con >= de 500 nan
print("\n verificamos que se elimino la columna \n")   
print(df_new.isnull().sum(axis=0)) #verificamos que se elimino deck

#________________________ELIMINAR POR FILAS_______________________________

df_age=df_new.dropna(subset=["age"],how="any",axis=0)#dropna elimina valores nulos, axis 0 indica filas
    #subset especifica la columna para aplicar el filtro y how="any" para decir que cualquiera
print("\n verificamos que se eliminaron las filas \n")   
print(df_age.isnull().sum(axis=0)) #verificamos que se eliminaron las filas

print("despues de filtrar eliminando tengo en total :", len(df_age))
```

# ejemplo2 REEMPLAZAR los valores NaN

```python
# trabajamos con df new (eliminamos deck)
print(df_new.isnull().sum(axis=0))
print()

#reemplazo age con el promedio
mean_age=int(df_new["age"].mean(axis=0))# obtener el promedio de age fila por fila
print("el promedio de edad es: ",mean_age ,"\n")

df_new["age"].fillna(mean_age,inplace=True) #fillna -> rellena los valores NaN

print("\n verificamos que se corrigieron las filas \n")  
print(df_new.isnull().sum(axis=0))

#reemplazo embark_town
#reemplazar con el dato que mas se repita
most_freq=df_new["embark_town"].value_counts(dropna=True).idxmax()
                                                #idxmax halla el valor que mas se repite
print("\n la ciudad que mas se repite es : ",most_freq ,"\n")

df_new["embark_town"].fillna(most_freq,inplace=True) #fillna -> rellena los valores NaN

print("\n verificamos que se corrigieron las filas en embark_town \n")  
print(df_new.isnull().sum(axis=0))

#reemplazo embarked
#reemplazar con el dato continuo
#ffill-> anterior
#bfill-> siguiente 

df_new["embarked"].fillna(method="ffill",inplace=True)
print("\n verificamos que se corrigieron las filas en embarked \n")  
print(df_new.isnull().sum(axis=0))



```
