# RECAPITULANDO EL EJEMPLO # COMPLETO

```python
#NOTA: 3 pasos : 
#    1:instanciar
#        1.1:definir el modelo
#    2:fit ->ajustar
#    3:predecir    

#para instanciar, utilice el hiperparametro del modelo como argumento.
#el hiperparametro es una opcion que REQUIERE CONFIGURACION HUMANA (AI SUPERVISADA!!!!!!!!)
#afecta MUCHO al rendimiento del modelo

#recapitulando el ejemplo

import pandas as pd
from IPython.core.display import HTML

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier #ver en la webpage : https://scikit-learn.org/stable/modules/tree.html#classification

#cargar los datos 
datos=load_breast_cancer()
#instanciar el estimador y ajustar los parametros
x_train,x_test,y_train,y_test=train_test_split(datos.data,datos.target)

#inicializacion del modelo de arbol(ramificacion) mediante entropia 
model=DecisionTreeClassifier(criterion="entropy")
```

# MODELO FIT (modelo de ajuste)
### Utilice el metodo de ajuste(FIT) con el estimador de instancia para el entrenamiento
### envie los datos  de entrenamiento y los datos de etiqueta juntos como argumento
###  al  algoritmo de aprendizaje  (IA) SUPERVISADO

```python
#OJO!!!!! AQUI ESTA APRENDIENDO :D
model.fit(x_train,y_train) #le mando los datos de train para que entrene y aprenda
#fit ajusta los datos para crear un modelo (regresion) que pueda predecir los datos

# predecir
#aqui le mando el test para ver si si aprendio
y_pred=model.predict(x_test) #le mando los datos de test para que ejecute y observar si si aprendio


#nota
#el estimador de instancia que ha completado el entrenamiento con el ajuste(fit) puede aplicarse
#con el metodo predecir
#predict() convierte los resultados estimados del modelo respecto  a los datos de prueba

y_pred
```

```python
#nota
#Es un valor estimado!!!!, por lo que el valor real de x_test puede variar
#para verificar podemos medir la presicion comparando los valores
```

```python
y_pred=model.predict(x_test)
pred_2d=y_pred.reshape(len(y_pred),1)#recordemos que y_pred son los datos predecidos
pred_2d
```

```python
#y_pred :son los datos predecidos f(x_test)
#y_test :son los resultados VERDADEROS!!!!!

y_test_2d=y_test.reshape(len(y_test),1)
y_test
```

```python
#para observar con facilidad los junto
df_pred=pd.DataFrame(pred_2d) #predecidos
df_true=pd.DataFrame(y_test_2d)#verdaderos
df_concat=pd.concat([df_pred,df_true],axis=1)
df_concat.columns=["pred","true"]
display(HTML(df_concat.to_html()))
```

# comprobamos la eficiencia de mi IA

```python
#busco los casos de exito de la IA, la prediccion fue correcta

df_exito=df_concat[df_concat["pred"]==df_concat["true"]]#muestra solo donde son iguales
display(HTML(df_exito.to_html()))
```

```python
exito=len(df_exito)
print(exito)
total_datos=len(df_concat)
print(total_datos)

eficiencia=exito/total_datos
print("la eficiencia de la IA es:",eficiencia*100,"%")
```
