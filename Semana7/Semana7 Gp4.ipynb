```python
#pasar de series a dataframe
import pandas as pd

s1=pd.Series([2,3,5,6,7],index=["a","b","c","d","e"])
s2=pd.Series([6,8,3,5,9],index=["a","b","c","d","e"])
print(s1)
print("__")
print(s2)

data={"serie1":s1,"serie2":s2}
dt=pd.DataFrame(data)

dt.index=["abrigos","jeans","zapatos","sombreros","relojes"]
dt.columns=["hombres","mujeres"]

dt.rename(index={"abrigos":"chaqueta"},inplace=True)
dt.rename(columns={"hombres":"niños","mujeres":"niñas"},inplace=True)

print(dt.shape) # filas,columnas
print(len(dt)) #retorna el numero de filas

print(dt.index)
print(dt.columns)
print(dt.values)








dt
```
