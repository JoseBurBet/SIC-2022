# verificar valores nulos

```python
import pandas as pd
import matplotlib.pyplot as plt

titanic=pd.read_csv("train.csv")

dt=titanic.head(7)
dt

edades=dt["Age"].head(7)
print(edades)

avg_age=dt["Age"].mean(axis=0) #axis=0 indica las filas , mean(saca el promedio de esas filas)  y dt["Age"] es la clumna de las edades
print(avg_age)

type(edades)

plt.bar(edades.index,edades.values)
plt.show()

print("_________________________-")

for i in range(len(edades.values)):
    #print(edades.values[i])
    if edades.values[i].isnull()==True:
        print(edades.index[i], "tiene un valor null")
        edades.values[i]=avg_age


edades.values[5]=avg_age
print(edades)

plt.bar(edades.index,edades.values)
plt.show()

#dt.isnull().sum()
```
